```python
Cars_url=[]
```

```python
for i in driver.find_elements_by_xpath("//img[@class='rg_i Q4LuWd']")[:150]:
    
    try:
        i.click()
        car=driver.find_element_by_xpath('//*[@id="Sva75c"]/div/div/div[3]/div[2]/c-wiz/div/div[1]/div[1]/div[2]/div[1]/a/img')
        Cars_url.append(car.get_attribute('src'))
        
    except NoSuchElementException as e:
        Cars_url.append("-")
    
    except ElementNotInteractableException as e:
        Cars_url.append("-")
        
    except StaleElementReferenceException as e:
        Cars_url.append("-")    
    
    except ElementClickInterceptedException as e:
        Cars_url.append("-")    
```
